```python
# Merge the two DataFrames on 'bodyId' to compare Q-values
comparing_q_values = pd.merge(q_values_computed[['bodyId','Q_value']], q_values_reneel[['bodyId','Q_value']], on='bodyId', suffixes=('_computed', '_reneel'))

# subtract the Q-values to see the difference
comparing_q_values['Q_value_difference'] = comparing_q_values['Q_value_computed'] - comparing_q_values['Q_value_reneel']

# show any non-zero differences
comparing_q_values[comparing_q_values['Q_value_difference'] != 0]
```
